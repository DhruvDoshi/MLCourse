# Mean, Median, Mode, and introducing NumPy

## Mean vs. Median

Let's create some fake income data, centered around 27,000 with a normal distribution and standard deviation of 15,000, with 10,000 data points. (We'll discuss those terms more later, if you're not familiar with them.)

Then, compute the mean (average) - it should be close to 27,000:

```python
import numpy as np

incomes = np.random.normal(27000, 15000, 10000)
# bell curve around the data around a certain point here 27000 with the standard deviation of 15000 and we have 10000 datapoints in the data set
np.mean(incomes)

# fun fact - Each time it will give a different output as this is the randum but here we will surely deviate around 27000 range 
```

We can segment the income data into 50 buckets, and plot it as a histogram:

```python
%matplotlib inline
import matplotlib.pyplot as plt
# Create the simple graph here in line and to actually display these graph within the notebook 
plt.hist(incomes, 5000)
#hist to create a histogram and 5000 will split up the data set into 50 different blocks -- 5000     descrete buckets of data sets
plt.show()
```

Now compute the median - since we have a nice, even distribution it too should be close to 27,000:

```python
np.median(incomes)
# majorly median and the main must be same but that's not mandatory
```

Now we'll add Jeff Bezos into the mix. Darn income inequality!

```python
incomes = np.append(incomes, [1000000000])
# Tinker the dataset with adding the huge value this must not change the median but will change the mean for sure
```

The median won't change much, but the mean does:

```python
np.median(incomes)
```

```python
np.mean(incomes)
# Here we can see the mean had been changed but median is better for the larger scene 
# Median is betterly placed to say what's happening in larger picture
```

## Mode

```python
ages = np.random.randint(18, high=90, size=1000)
# Range of ages from 18 to 90 for 1000 people 
ages
```

```python
from scipy import stats
stats.mode(ages)
# This is completely random, each time we fire it out we will get the newer values and outputs
```
